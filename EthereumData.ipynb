```python
# Basic settings and data import

import numpy as np
import pandas as pd
from pylab import mpl,plt
plt.style.use('seaborn')
mpl.rcParams['font.family'] = 'serif'
%matplotlib inline
```

```python
# Import CSV files with Ethereum price data from Yahoo finance, 2021/03/01-2022/03/01

filename_daily = 'C:/Users/admin/Desktop/OVGU Magdeburg/3. Semester MA/Proseminar/Literatur/Ethereum_daily.csv'
filename_weekly = 'C:/Users/admin/Desktop/OVGU Magdeburg/3. Semester MA/Proseminar/Literatur/Ethereum_weekly.csv'
```

```python
# Create data frame of daily adjusted close prices

f = open(filename_daily, 'r')
data_daily = pd.read_csv(filename_daily, index_col=0, parse_dates=True)
daily_close = data_daily['Adj Close']
```

```python
# Plot daily returns of one year period (used in technical report)

daily_close.plot(figsize=(10,6), grid=('false'));
```

```python
# Calculate percentage change of daily prices

daily_change = daily_close.pct_change().round(3)
```

```python
# Get maximum and minimum percentage changes

maxdaily = daily_change.max()
mindaily = daily_change.min()
```

```python
# Print maximum gain

print(maxdaily)
```

```python
# Print maximum loss

print(mindaily)
```

```python
# Redo steps with weekly data

fw = open(filename_weekly, 'r')
data_weekly = pd.read_csv(filename_weekly, index_col=0, parse_dates=True)
weekly_close = data_weekly['Adj Close']
weekly_change = weekly_close.pct_change().round(3)
maxweekly = weekly_change.max()
minweekly = weekly_change.min()
```
